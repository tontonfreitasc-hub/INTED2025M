2 - Crie um algoritimo que leia 2 números e imprima na tela o resultado das operações acima.

```python
#Entradas
numero1 = int(input("digite o primeiro número: "))
numero2 = int(input("digite o segundo número: "))

#Processamento
soma = numero1 + numero2
subt = numero1 - numero2
mult =  numero1 *numero2
div = numero1 / numero2
resto = numero1 % numero2
expo = numero1 ** numero2

#Saída
print(f"A soma é: {soma}")
print(f"A subtração é: {subt}")
print(f"A divisão é: {div:.2f}")
print(f"A multiplicação é: {mult}")
print(f"O resto é: {resto}")
print(f"A exponenciação é: {expo:.2e}")
```

3 - Crie um algorítimo que receba o nome do usuário e 3 notas, depois calcule a média na variável (média) e imprima na tela a seguinte frase:

O aluno (nome) está com (média) de média!

```python
# Entrada de dados
nome = input("Digite o nome do aluno: ")
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segunda nota: "))
nota3 = float(input("Digite a terceira nota: "))

# Cálculo da média
media = (nota1 + nota2 + nota3) / 3

# Saída de dados
print(f"O aluno {nome} está com {media:.2f} de média!")
```
